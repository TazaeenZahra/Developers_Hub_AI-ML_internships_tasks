### Visualize the dataset:

○ Create a scatter plot to show relationships between features.

○ Use histograms to show value distributions.

○ Use box plots to identify outliers.

● Use matplotlib and seaborn for plotting.


```python
sns.scatterplot(x = 'sepal_length',y= 'petal_length',data = data,hue='species')

plt.title('Scatter plot of sepal length vs petal length')
plt.show()
```

```python
data.hist()

plt.suptitle("Feature Distributions")
plt.show()
```

```python
sns.histplot(data['petal_width'])

plt.title('petal_width histogram')
plt.show()
```

```python
sns.histplot(data['petal_length'])

plt.title('petal_length histogram')
plt.show()
```

```python
sns.histplot(data['sepal_length'])

plt.title('sepal_length histogram')
plt.show()
```

```python
sns.histplot(data['sepal_width'])

plt.title('sepal_width histogram')
plt.show()
```

```python
sns.boxplot(data=df)

plt.title("Box Plot of Iris Features")


plt.show()
```
